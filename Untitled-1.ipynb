```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("processed_data2.csv")
df.head()
```

```python
label_encoder = LabelEncoder()
df['battery_size'] = label_encoder.fit_transform(df['battery_size'])
df['os_type'] = label_encoder.fit_transform(df['os_type'])
df['display_type'] = label_encoder.fit_transform(df['display_type'])
```

```python
X = df[["ram", "storage", "weight", "battery_size", "os_type", "display_type"]]
Y = df["price_usd"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

coefficients = model.coef_

# AfiÈ™area rezultatelor
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
print(f"Coefficient for os_type: {coefficients[4]}")
print(f"Coefficient for weight: {coefficients[2]}")
print(f"Coefficient for battery_size: {coefficients[3]}")
print(f"Coefficient for ram: {coefficients[0]}")
print(f"Coefficient for display: {coefficients[5]}")
# print(label_encoder.classes_)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

plt.scatter(Y_test, Y_pred, color='blue', label='Predicted vs Actual')

plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--', label='Perfect Prediction')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression: Actual vs Predicted')
plt.legend()


plt.show()
```
